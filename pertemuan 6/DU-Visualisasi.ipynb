```python
import matplotlib.pyplot as plt 
import numpy as np

negara=('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 9750000, 273500000, 1250000, 2655000)
kor_x = np.arange(len(negara))
plt.bar(kor_x, populasi, tick_label=negara)
plt.xticks(rotation = 90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

negara=('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 9750000, 273500000, 1250000, 2655000)

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by='Population', inplace = True)

kor_x = np.arange(len(df))
plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation = 90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show()
```

```python
## Chalenge besar ke kecil
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

negara=('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 9750000, 273500000, 1250000, 2655000)

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by='Population', inplace=True, ascending=False)

kor_x = np.arange(len(df))
plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation = 90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show()
```

```python
## cara higlight
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

negara=('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 9750000, 273500000, 1250000, 2655000)

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by='Population', inplace = True)

kor_x = np.arange(len(df))
warna = ['#FFE4C4' for _ in range(len(df))]
warna [-3] = '#808000' 

plt.bar(kor_x, df['Population'], tick_label=df['Country'], color=warna)
plt.xticks(rotation = 90)
plt.ylabel('Populasi (Juta)')
plt.title('Negara ke 3 Terbesar')
plt.show()
```
