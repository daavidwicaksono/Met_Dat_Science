## 2. Pie Chart
Disini saya menggunakan pie chart, karena dalam menampilkan datanya menggunakan autopct yang memberi kemungkinan terambilnya masing masing permen.

```python
import matplotlib.pyplot as plt 

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]
warna = ('#FDF5E6', '#D2691E', '#00BFFF', '#FF00FF', '#FFD700')
highlight = (0,0.2,0,0.0,0)

plt.title ('Pie Chart Kemungkinan Terambilnya Permen')
plt.pie(jumlah_permen, labels = nama_permen, autopct = '%1.1f%%', colors = warna, explode=highlight, shadow= True)
        
plt.show
```

Pada hasi pie chart diatas, menghasilkan presentase kemunculan permen kopiko yaitu 15% atau 3/20.

## 3.  Bar chart
Disini saya menggunakan bar chart, karena bar chart bisa menunjukkan jumlah penjualan paling sedikit dan paling banyak secara berurutan.

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd


datapenjualan_makananpenutup = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 
                                'Brownies', 'Puding Strawberry', 'Puding Coklat','Ice Cream Nutela', 
                                'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jmlJual = (14, 5, 12, 19,  6, 8, 12, 9,  10, 17, 2, 9, 13)

df = pd.DataFrame({'Penjualan':datapenjualan_makananpenutup, 'Jumlah_Terjual':jmlJual})
df.sort_values(by='Jumlah_Terjual', inplace = True)

kor_x = np.arange(len(df))
warna = ['#1E90FF' for _ in range(len(df))]
warna [-11] = '#DC143C' 
warna [-12] = '#DC143C' 
warna [-13] = '#DC143C'

plt.bar(kor_x, df['Jumlah_Terjual'], tick_label=df['Penjualan'], color=warna)
plt.xticks(rotation = 90)
plt.xlabel('Nama Makanan')
plt.ylabel('Jumlah Yang Terjual')
plt.title('Daftar Makanan Penutup')
plt.show()
```

1. Visualisasi popularitas relatif dari makanan penutup.
dari grafik diatas dapat menghasilkan Donat, Kue Coklat-Keju, dam Ice Cream menjadi makanan penutup favorit
2. Tiga makanan yang harus disingkirkan
dari grafik diatas dapat menghasilkan Kue Wajik, Pastel, dan Puding Vanila menjadi makanan yang harus disingkirkan
